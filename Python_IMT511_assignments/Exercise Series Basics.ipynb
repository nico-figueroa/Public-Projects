```python
import pandas as pd
import numpy as np
```

```python
nums = pd.Series([5,2,6,2,1,6])
print(nums)
```

```python
print(nums.index)
```

```python
nums.index = ['a','b','c','d','e','f']
print(nums)
```

```python
monotonic = pd.Series(range(1,7))
print(monotonic)
```

```python
print(nums + monotonic)
# The reason why the result is NaN is that there are no matching indices between the two series
```

```python
remainder_3 = nums % 3
print(remainder_3)
```

```python
print(nums.median())
```

```python
print(nums.idxmin())
```

```python
divisible_by_3 = pd.Series(nums % 3 == 0)
print(divisible_by_3)
```

```python
print(remainder_3.any())
```
