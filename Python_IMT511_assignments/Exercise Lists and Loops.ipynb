```python
# The count_odds function takes a list of numbers as an argument and
# returns how many elements in the list are odd

def count_odds(number_list):
    
    total_odds = 0 # initialize counter
    
    for number in number_list: 
        if number % 2 != 0:
            total_odds = total_odds + 1

    return total_odds

print(count_odds(phone_numbers))
```

```python
# The get_odd_values function takes a list of numbers as an argument and
# returns a list of the odd elements

def get_odd_values(number_list):
    
    odd_values_list = [] # create initial list
    
    for number in number_list: 
        if number % 2 != 0:
            odd_values_list.append(number) # append the odd value found to the list

    return odd_values_list

print(get_odd_values(phone_numbers))
```

```python
# The get_odd_indices function takes a list of numbers as an argument and
# returns a list of the indices of the elements that have odd values

def get_odd_indices(number_list):
    
    odd_values_indices = [] # create initial list
    
    for index in range(len(number_list)): 
        if number_list[index] % 2 != 0:
            odd_values_indices.append(index) # append the index number to the list of indices

    return odd_values_indices

print(get_odd_indices(phone_numbers))
```

```python
# The rotate_left function takes a non-empty list of elements as an argument and
# rotates it by moving each element 1 spot to the left, where the first element becomes the last

def rotate_left(input_list):
    for index in range(1,len(input_list)):
        input_list[index], input_list[index-1] = input_list[index-1], input_list[index]
        
my_list = ['a','b','c','d']
rotate_left(my_list)
print(my_list)
```

```python
# The function make_clapping_song takes a name as an argument
# and returns a nested list with each letter of the name
# progressively being replaced by an asterisk

def make_clapping_song(name):
    clapping_song = [] # song placeholder creation
    
    for position_index in range(len(name)+1):
        clapping_song.append(list(name)) # add name lists times the number of letters of the name
        
        for list_index in range(position_index):
            clapping_song[position_index][list_index] = '*' # replace an asterisk in every position based on row and column
    
    return clapping_song
```

```python
bingo_song = make_clapping_song("BINGO")
print(bingo_song)
```

```python
for position_index in range(len(bingo_song)):
    for list_index in range(position_index):
        bingo_song[position_index][list_index] = "*clap*"

print(bingo_song)
```

```python
# The rotate_text function takes a message and a number of places to shift letters as arguments and
# shifts the letters in the message the designated quantity. The function encrypts when using a positive
# number of places and descrypts when using a negative number

def rotate_text(message,shift_places):
    
    import string
    
    message = message.upper()
    ciphered_message = '' # initialization of new message variable
    
    for letter in message:
        letter_index = string.ascii_uppercase.find(letter) # find the index of each letter of the message in the alphabet

        if letter_index != -1 and letter_index-shift_places < 26: # confirm that index does not exceed alphabet length
            ciphered_message = ciphered_message + string.ascii_uppercase[(letter_index-shift_places)]
        elif letter_index != -1 and letter_index-shift_places >= 26: # if alphabet length is exceeded, return to start of alphabet and then shift
            ciphered_message = ciphered_message + string.ascii_uppercase[(letter_index-shift_places-26)]
        else: # used to add spaces to output message
            ciphered_message = ciphered_message + ' '

    return ciphered_message

print(rotate_text("THE QUICK BROWN FOX JUMPS OVER THE LAZY DOG",3))
print(rotate_text("QEB NRFZH YOLTK CLU GRJMP LSBO QEB IXWV ALD",-3))
```
