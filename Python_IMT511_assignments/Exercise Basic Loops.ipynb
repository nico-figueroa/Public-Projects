```python
# The is_prime function takes a whole number as an argument and returns
# whether or not that number is prime (divisible by only 1 and itself).
# The number argument passed to the function must be 2 or larger

def is_prime(number):
    count = 0 # initialize count of divisibles

    for iteration in range(2,number+1):   
        if number % iteration == 0:
            count = count + 1 # increase number of divisibles found
        
    if count == 1: # prime numbers can only be divided by themselves
        conclusion = True
    else:
        conclusion = False
    
    return conclusion
```

```python
print(is_prime(97))
```

```python
# The print_primes function prints out all the prime numbers that are
# less than or equal to the number in the argument

def print_primes(number):
    for iteration in range(0,number+1): # add 1 to be inclusive of the number entered
        if is_prime(iteration): # find the primes in the iteration
            print(iteration, end = ", ")

print_primes(100)
```

```python
# The percent_vowels function takes a string as an argument
# and returns the percent of vowels (a, e, i ,o u or y) that
# the argument contains

def percent_vowels(word):
    tally = 0 # initialize count of vowels present
    
    for index in range(len(word)):
        if word[index] in ['a','e','i','o','u','y']:
            tally = tally + 1
        
    ratio = (tally / len(word)) 

    return ratio
```

```python
print(percent_vowels("computer"))
```
