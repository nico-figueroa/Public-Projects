```python
import random
LIST_SIZE = 10000
numbers = random.sample([number for number in range(0,LIST_SIZE**2)],LIST_SIZE)
targets = random.choices(numbers,k=1000)
```

```python
# The linear search function takes a list and a target value and returns the index of the 
# target values within the list or -1 if the targets are not found.

def linear_search(search_list,target):
    
    for index in range(len(search_list)): # identify the list positions
    
        if search_list[index] == target: # look for the target in every position of the list
            return index # return the position value of the target
    
    return -1 # did not find the target
```

```python
import time
print(time.time())
```

```python
start_time = time.time()
for target in targets:
    linear_search(numbers,target)
end_time = time.time()
print(end_time-start_time)
```

```python
# Function provided

def binary_search(a_list, target):
    """Searches the given SORTED list for the given target value.
       Returns the index of the target, or -1 if target not in the list"""

    start_index = 0 # initial goalposts
    end_index = len(a_list)-1

    while start_index <= end_index:  # at least one thing to look at
        middle_index = (start_index + end_index) // 2  # middle (integer) index

        if(a_list[middle_index] == target):
            return middle_index  # found the item!
        elif target > a_list[middle_index]:
            start_index = middle_index+1  # move goalpost
        else:
            end_index = middle_index-1  # move goalpost

    return -1  # did not find the item
```

```python
start_time = time.time()
for target in targets:
    binary_search(numbers,target)
end_time = time.time()
print(end_time-start_time)
```

```python
# Function from text book "Intro to Programming section 7.4 which sorts the list in place.
# (https://infx511.github.io/searching.html#linear-search)

def selection_sort(a_list):

    for i in range(len(a_list)):  # go through each spot in the list
        # Do a "king-of-the-hill" search of the remaining items
        selected_index = i
        for j in range(i, len(a_list)):
            if(a_list[j] < a_list[selected_index]):
                selected_index = j

        # swap smallest into place (multi-assignment!)
        a_list[i], a_list[selected_index] = a_list[selected_index], a_list[i]
```

```python
start_time = time.time()
selection_sort(numbers)
end_time = time.time()
print(end_time-start_time)
```
