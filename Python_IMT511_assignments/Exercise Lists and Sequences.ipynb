```python
# The get_middle_value function takes a list as an argument and 
# returns the value in the middle position of the list which is
# the exact middle for a list with odd number of items and the
# round down item for a list with even number of items

def get_middle_value(a_list):

    if len(a_list) % 2 == 0: # check if its even
        middle_value = a_list[int((len(a_list)/2)-1)] # the round down value is assigned because the list is even
    else:
        middle_value = a_list[int((len(a_list)/2)-0.5)] # the exact middle is assigned because the list is odd

    return middle_value 

get_middle_value([1,2,3,4])
get_middle_value([1,2,3,4,5])
```
