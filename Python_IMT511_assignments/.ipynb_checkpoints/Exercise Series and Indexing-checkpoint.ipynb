```python
import pandas as pd
import numpy as np
```

```python
phone_numbers = pd.Series([5, 5, 5, 8, 6, 7, 5, 3, 0, 9])
```

```python
print(phone_numbers[5])
```

```python
prefix = phone_numbers[[3,4,5]]
print(prefix)
```

```python
not_5 = phone_numbers[phone_numbers != 5]
print(not_5)
```

```python
small_nums = phone_numbers[phone_numbers < phone_numbers.median()]
print(small_nums)
```

```python
phone_numbers[phone_numbers > 7] = 7
print(phone_numbers)
```

```python
phone_numbers[phone_numbers.index % 3 == 0] = 511
print(phone_numbers)
```

```python
words = pd.Series(["the", "quick", "brown", "fox", "jumped", "over", "lazy", "dog"])
print(words)
```

```python
capital_words = words.str.capitalize()
print(capital_words)
```

```python
import string
letters = string.ascii_lowercase
letters_tuple = tuple(letters)
a_letter_index = letters.find('a')
f_letter_index = letters.find('f')
print(letters)
print(letters_tuple)
print(a_letter_index)
print(f_letter_index)
print(letters_tuple[a_letter_index:f_letter_index+1])
```

```python
words_a_f = words[words.str > 'a' and words.str.lower() < 'f']
print(words_a_f)
```
