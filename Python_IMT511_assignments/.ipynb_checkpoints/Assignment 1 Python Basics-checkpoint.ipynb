```python
import math
def compound_interest(principal,annual_interest_rate,years):
    resulting_total_balance = principal * math.exp(annual_interest_rate * years)
    return resulting_total_balance
compound_interest(1000,0.06,5)
```

```python
def print_earnings(principal,annual_interest_rate,years):
    total_value = compound_interest(principal,annual_interest_rate,years)
    interest_earned = total_value - principal
    print("Initial principal: $" + str(round(principal,2))) 
    print("Annual interest rate: %" + str(round(annual_interest_rate*100,1))) 
    print("Interest earned in 5 years: $" + str(round(interest_earned,2)))
    print("Total value after 5 years: $" + str(round(total_value,2)))
print_earnings(1000,0.06,5)
```

```python
def value_of_change(quarters=0,dimes=0,nickels=0,pennies=0):
    total_cents = (quarters * 25) + (dimes * 10) + (nickels * 5) + pennies
    total_dollar_value = total_cents / 100
    return total_dollar_value
value_of_change(3,2)
```

```python
def consolidate_change(quarters=0,dimes=0,nickels=0,pennies=0):
    total_amount = value_of_change(quarters,dimes,nickels,pennies)
    total_pennies = total_amount * 100
    dollar_amount = int(total_pennies // 100)
    quarters_amount = int((total_pennies - (dollar_amount * 100)) // 25)
    dimes_amount = int((total_pennies - (dollar_amount * 100) - (quarters_amount * 25)) // 10)
    nickels_amount = int((total_pennies - (dollar_amount * 100) - (quarters_amount * 25) - (dimes_amount * 10)) // 5)
    pennies_amount = int((total_pennies - (dollar_amount * 100) - (quarters_amount * 25) - (dimes_amount * 10) - (nickels_amount * 5)) // 5)
    print("Number of dollars: " + str(dollar_amount))
    print("Number of quarters: " + str(quarters_amount))
    print("Number of dimes: " + str(dimes_amount))
    print("Number of nickels: " + str(nickels_amount))
    print("Number of pennies: " + str(pennies_amount))
    print("Total amount: $" + str(total_amount))
consolidate_change(10,9,8,7)
```

```python
consolidate_change(47,56,89,54)
```

```python
from dateutil.relativedelta import relativedelta
def when_can_vote_us(birth_date):
    majority_of_age = birth_date + relativedelta(years=+18)
    return majority_of_age
```

```python
print(when_can_vote_us(date.today()))
```

```python
years_old = relativedelta(date.today(),date(1982,1,17)).years
last_birthday = date(1982,1,17) + relativedelta(years=+(years_old-1))
next_birthday = date(1982,1,17) + relativedelta(years=+(years_old+1))
days_till_next_birthday = next_birthday - date.today()
print(years_old)
print(next_birthday)
print(days_till_next_birthday.days)
```

```python
def make_birthday_announcement(name,date_of_birth):
    years_old = relativedelta(date.today(),date_of_birth).years
    next_birthday = date_of_birth + relativedelta(years=+(years_old + 1))
    days_till_next_birthday = next_birthday - date.today()
    birthday_announcement = name + "'s birthday is in " + str(days_till_next_birthday.days) + " days"
    return birthday_announcement
```

```python
print(make_birthday_announcement(my_name,date(1982,1,17)))
```
