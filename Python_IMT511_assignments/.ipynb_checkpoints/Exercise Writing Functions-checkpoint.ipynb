```python
def add_three(original_value):
    increased_value = original_value + 3
    return increased_value
```

```python
ten = add_three(7)
print(ten)
```

```python
add_three("7")
```

```python
def imperial_to_metric(number_of_feet,number_of_inches):
    meters = number_of_feet*12*0.0254 + number_of_inches*0.0254
    return meters
```

```python
height_in_meters = imperial_to_metric(5,10)
print(height_in_meters)
```

```python
def compare_str_length(string_one, string_two):
    string_difference = abs(len(string_one) - len(string_two))
    answer = "The difference in string lengths is " + str(string_difference)
    return answer
```

```python
compare_str_length("Nicolas","Figueroa")
```

```python
def fraction_str(numerator=1,denominator=1):
    string_fraction = str(numerator) + "/" + str(denominator)
    return string_fraction
```

```python
print(fraction_str(numerator=5,denominator=11))
print(fraction_str(denominator=11,numerator=5))
```

```python
print(fraction_str(denominator=3))
```

```python
print(fraction_str(11,5))
```
