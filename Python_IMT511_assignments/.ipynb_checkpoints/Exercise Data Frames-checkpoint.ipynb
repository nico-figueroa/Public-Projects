```python
import pandas as pd
import numpy as np
```

```python
employee_names = pd.Series(['Employee '+str(number) for number in range(1,51)])
print(employee_names.head(3))
print(employee_names.tail(3))
```

```python
random_generator = np.random.default_rng(511)

salaries_2020 = pd.Series(random_generator.integers(low=40000,high=60000,size=50))
print(salaries_2020.head(5))
```

```python
salaries_df = pd.DataFrame({'employee': employee_names, 'salary_2020':salaries_2020})
```

```python
print(salaries_df.head(20))
```

```python
salaries_df = salaries_df.assign(adjustment_2020=random_generator.integers(low=-500,high=2000,size=50))
print(salaries_df.head(5))
```

```python
salaries_df['salary_2021']= salaries_df['salary_2020']+salaries_df['adjustment_2020']
print(salaries_df.head(5))
```

```python
salaries_df['salary_2022'] = salaries_df.salary_2021
print(salaries_df.head(5))
```

```python
salaries_df['salary_2022'] = salaries_df.salary_2022*1.02
print(salaries_df.head(5))
```

```python
salaries_only_df = salaries_df[['salary_2020','salary_2021','salary_2022']]
print(salaries_only_df.head(5))
```

```python
salary_stats_df = salaries_only_df.describe()
print(salary_stats_df)
```

```python
median_salaries = salaries_only_df.median()
print(median_salaries)
```

```python
paycut_df = salaries_df[salaries_df.adjustment_2020 < 0]
print(paycut_df)
```

```python
employee_15_salary_2021 = int(salaries_df.loc[14,'salary_2021'])
print(employee_15_salary_2021)
```

```python
biggest_raise_employee = salaries_df.iloc[salaries_df['adjustment_2020'].idxmax()]
print(biggest_raise_employee)
```

```python
top_employees = salaries_df[(salaries_only_df >= salary_stats_df.loc['75%']).any(axis=1)]
print(top_employees)
```
