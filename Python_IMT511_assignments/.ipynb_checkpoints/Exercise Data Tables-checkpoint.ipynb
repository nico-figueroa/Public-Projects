```python
# The create_game_dict function receives a home team name and score, and a visiting team name and score,
# and returns a dictionary of these entries

def create_game_dict(home_team,visiting_team,home_score,visiting_score):
    
    game_dict = {'home':home_team,'opp':visiting_team,'home_score':home_score,'opp_score':visiting_score}

    return game_dict
```

```python
rutgers_game = create_game_dict('UW','Rutgers',48,13)
print(rutgers_game)
```

```python
idaho_game = create_game_dict('UW','Idaho',59,14)
portland_game = create_game_dict('UW','Portland State',41,3)
uw_game = create_game_dict('Arizona','UW',28,35)

season2016 = [rutgers_game,idaho_game,portland_game,uw_game]

print(season2016)
```

```python
print(season2016[2]['opp'])
```

```python
for game in season2016:
    print(game['home']+" vs. "+game['opp'])
```

```python
home_won_count = 0
for game in season2016:
    if game['home_score'] > game['opp_score']:
        home_won_count = home_won_count+1

print(home_won_count)
```

```python
home_points = 0
for game in season2016:
    if game['home_score'] > game['opp_score']:
        
        home_points = home_points + (game['home_score']-game['opp_score'])
        
home_point_spread = home_points / home_won_count

print(home_point_spread)
```

```python
# The analyze_full_season function reads the lines of the huskies_2016.csv file
# and obtains the home_won_count and home_point_spread calculations for the 
# whole season included in the same

def analyze_full_season():
   
    with open('data/huskies_2016.csv') as file:
        data_list = file.readlines()

    game_list = []
    
    for item in data_list[1:]:

        records = tuple((item.strip()).split(sep=','))
        game_list.append(records)

    season2016_full = [create_game_dict(game[1],game[2],int(game[3]),int(game[4])) for game in game_list]

    print("Games List:")
    for game in season2016_full:
        print(game['home']+" "+str(game['home_score'])+" vs. "+game['opp']+" "+str(game['opp_score']))

    home_won_count = 0
    
    for game in season2016_full:
        if game['home_score'] > game['opp_score']:
            home_won_count = home_won_count+1

    print("\n" + "Home Wins:", home_won_count)

    home_points = 0
    
    for game in season2016_full:
        if game['home_score'] > game['opp_score']:
        
            home_points = home_points + (game['home_score']-game['opp_score'])
        
    home_point_spread = home_points / home_won_count

    print("\n" + "Home Point Spread:", home_point_spread)

analyze_full_season()
```
