```python
import pandas as pd
import numpy as np
```

```python
temp_data_df = pd.read_csv('data/Annual_Surface_Temperature_Change.csv')
```

```python
temp_data_df.head(5)
```

```python
print(list(temp_data_df.columns))
```

```python
temp_data_df = temp_data_df.drop(['ObjectId', 'Indicator', 'Unit', 'Source', 'CTS_Code', 'CTS_Name', 'CTS_Full_Descriptor'], axis='columns')
print(list(temp_data_df.columns))
```

```python
temp_data_df.columns = [column.removeprefix('F') for column in list(temp_data_df.columns)]
temp_data_df.head(5)
```

```python
temp_data_df = temp_data_df.drop(temp_data_df.loc[:,'1961':'1999'].columns, axis='columns')
temp_data_df.head(5)
```

```python
usa_data_series = temp_data_df.iloc[temp_data_df.index[temp_data_df['ISO3'] == 'USA'][0]]
print(usa_data_series)
```

```python
average_temp_change_usa = usa_data_series[3:].mean()
print(average_temp_change_usa)
```

```python
hottest_year_usa = usa_data_series[3:].idxmax()
print("Hottest year:",hottest_year_usa,"Temperature Difference:",usa_data_series[hottest_year_usa])
```

```python
avg_2010_to_2012 = temp_data_df[['2010','2011','2012']].mean(axis="columns")
print(avg_2010_to_2012)
```

```python
years_2000_2022_list = [str(year) for year in range(2000,2023)]
three_year_avg_df = temp_data_df.assign(
    avg2013_2015 = temp_data_df[["2013","2014","2015"]].mean(axis="columns"),
    avg2016_2018 = temp_data_df[["2016","2017","2018"]].mean(axis="columns"),
    avg2019_2022 = temp_data_df[["2019","2020","2021","2022"]].mean(axis="columns")
    ).drop(years_2000_2022_list,axis="columns")
three_year_avg_df.head(5)
```

```python
print(three_year_avg_df.loc[(three_year_avg_df['avg2019_2022']-three_year_avg_df['avg2016_2018']) > (three_year_avg_df['avg2016_2018']-three_year_avg_df['avg2013_2015'])]['Country'])
```

```python
global_averages_by_year = temp_data_df[years_2000_2022_list].mean(axis="index")
print(global_averages_by_year)
```

```python
import matplotlib.pyplot as plt
```

```python
# calculate the trendline
m, b = np.polyfit(np.arange(2000, 2023), global_averages_by_year, 1)
line = [(m * x) + b for x in np.arange(2000, 2023)]
plt.plot(line, linewidth = 3, color = 'Red') # include trendline in the plot
plt.plot(global_averages_by_year, scalex=True, linewidth = 2)
plt.tight_layout() # avoid label clipping
plt.grid(True) # show grid to enhance readability
plt.title("Average Temperature Difference per Year", color = 'Blue', fontsize=16)
plt.xlabel("Year", color = 'Blue', labelpad = 10)
plt.ylabel("Average Temperature Difference\n (°C)", color = 'Blue')
plt.xticks(rotation='vertical') # ensure dates can be read
plt.show() # clean output
```

```python
countries_df = pd.read_csv('data/iso_country_codes.csv')
print(countries_df.head(10))
```

```python
countries_df = countries_df.drop(['name','ISO2'], axis='columns')
```

```python
temp_data_region_df = temp_data_df.merge(countries_df, how='left', on='ISO3', validate='1:1')
```

```python
temp_data_region_df.head(5)
```

```python
temp_data_by_region = temp_data_region_df.groupby('region')
print(temp_data_by_region.agg(['count'])['Country'])
```

```python
print((temp_data_by_region.agg('max', numeric_only=True)['2022']).idxmax())
```

```python
print((temp_data_by_region.agg('mean', numeric_only=True)['2022']-temp_data_by_region.agg('mean', numeric_only=True)['2000']))
```

```python
# The function number_years_warmer receives two region names as arguments and returns
# the number of years in which the first region was on average "warmer" than the
# second region.

temp_data_by_region_t = temp_data_by_region.agg('mean', numeric_only=True).T

def number_years_warmer(first_region_name,second_region_name):

    number_years_warmer = len(temp_data_by_region_t.loc[temp_data_by_region_t[first_region_name] > temp_data_by_region_t[second_region_name]])

    return number_years_warmer 

print(number_years_warmer("Americas","Africa"))
```

```python
# calculate the trendline
m, b = np.polyfit(np.arange(2000, 2023), temp_data_by_region_t, 1)
line = [(m * x) + b for x in np.arange(2000, 2023)]
plt.figure(figsize=(9,6)) # control the plot size
plt.plot(line, linewidth = 2) # include trendline in the plot
plt.gca().set_prop_cycle(None) # reset the color cycles so that both graphs match colors
plt.plot(temp_data_by_region_t, scalex=True, scaley=True, linewidth = 0.5)
plt.tight_layout() # avoid label clipping
plt.grid(True) # show grid to enhance readability
plt.title("Regional Average Temperature Difference per Year", color = 'Blue', fontsize=16)
plt.xlabel("Year", color = 'Blue', labelpad = 10)
plt.ylabel("Average Temperature Difference\n (°C)", color = 'Blue')
plt.legend(temp_data_by_region_t.columns)
plt.xticks(rotation='vertical') # ensure dates can be read
plt.show() # clean output
```

```python
# What the graph shows is that the average temperature difference has increased in all regions between 
# the year 2000 and 2022. It also shows that Europe has the increases are proportional to the baseline
# starting point for most regions, except for Africa and Asia who swaped places in their average
# temperature difference, where Asia is now higher. Finally, the rates of increase vary between the
# different regions, with Europe, Asia and Oceania having the highest and comparable growth rates.
```

```python
import geopandas as gpd
geo_data = gpd.read_file('data/World_Countries_Generalized.geojson')
geo_data.head(5)
```

```python
geo_data.plot()
```

```python
mean_temp_change_df = temp_data_df[['Country','ISO2','ISO3']].copy()
mean_temp_change_df = mean_temp_change_df.assign(Average = temp_data_df.loc[:,'2012':'2022'].mean(axis='columns'))
mean_temp_change_df.head(5)
```

```python
geo_temp_data = geo_data.merge(mean_temp_change_df, left_on='ISO', right_on='ISO2', how='left')
```

```python
import matplotlib.colors as colors

# Set the range values from negative to positive
vmin = geo_temp_data['Average'].min()
vmax = geo_temp_data['Average'].max()

# Ensure numeric values in the range limits
limit = max(abs(vmin), abs(vmax))

ax = geo_temp_data.plot(column = "Average", 
                   legend = True, 
                   missing_kwds = {'color': "grey"}, 
                   cmap = "coolwarm", 
                   legend_kwds = {"shrink":0.5},
                   norm=colors.TwoSlopeNorm(vmin=-limit,
                                            vcenter=0,
                                            vmax=limit)
                  )

fig = ax.get_figure()
fig.set_size_inches(16,10) # allow size modification of the entire map
cbar = fig.axes[-1]  # Allow label rotation for the degrees symbol
cbar.set_ylabel("°C", rotation=0, labelpad=10, fontsize=15)
ax.set_axis_off() # remove axis and edges
plt.title("Average Temperature Change per Country", fontsize=20)
plt.show() # clean output
```
