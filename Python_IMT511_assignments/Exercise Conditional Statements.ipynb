```python
# The describe_length_difference function takes two strings as input arguments
# and evaluates the lengths of the strings to determine if one is bigger than the
# other or if they both have the same length and prints a statement with the result

def describe_length_difference(first_string,second_string):
    
    string_difference = abs(len(first_string) - len(second_string)) # determine the absolute difference between strings
    
    if len(first_string) > len(second_string): # perform comparisons for the string lengths and print respective message
        print("The first string is longer by " + str(string_difference) + " characters")
    elif len(first_string) < len(second_string):
        print("The second string is longer by " + str(string_difference) + " characters")
    else:
        print("The strings are the same length!")        
```

```python
describe_length_difference("Nico","Figueroa")
describe_length_difference("Programming","Python")
describe_length_difference("University","Washington")
```

```python
# The lucky_sum function takes three numbers as arguments and returns the sum total of those numbers
# ignoring any arguments that are either four or thirteen which are considered unlucky

def lucky_sum(first_number,second_number,third_number):
    
    if (first_number == 4 or first_number == 13): # determine if the first number is unlucky
        first_number = 0
    if (second_number == 4 or second_number == 13): # determine if the second number is unlucky
        second_number = 0
    if (third_number == 4 or third_number == 13): # determine if the third number is unlucky
        third_number = 0

    resulting_value = first_number + second_number + third_number # perform the sum of the converted numbers

    return resulting_value # return the result

# perform tests for the function trying lucky and unlucky numbers in different argument positions

print(lucky_sum(1,2,3)) # Expected 6
print(lucky_sum(4,5,6)) # Expected 11
print(lucky_sum(5,4,6)) # Expected 11
print(lucky_sum(5,6,4)) # Expected 11
print(lucky_sum(13,14,15)) # Expected 29
print(lucky_sum(14,13,15)) # Expected 29
print(lucky_sum(14,15,13)) # Expected 29
print(lucky_sum(4,13,14)) # Expected 14
print(lucky_sum(14,4,13)) # Expected 14
print(lucky_sum(4,14,13)) # Expected 14
```

```python
# The desired_bedrooms function takes an expected number of occupants, 
# a boolean value for whether the occupants are partnered or married
# and a combined annual income number and returns the number of bedrooms that
# the user should look for in an apartment

def desired_bedrooms(expected_occupants,partnered,annual_income):

    if expected_occupants <= 3: # branch on the left of the decision tree
        if annual_income < 40000: # continue going left
            if partnered: # left endpoint
                number_of_bedrooms_needed = 1
            else: # alternative income branch
                number_of_bedrooms_needed = 2
        else:
            number_of_bedrooms_needed = 3
    else: # take the branch on the right of the decision tree
        if not partnered: # continue right
            if annual_income >= 80000: # right endpoint
                number_of_bedrooms_needed = 4
            else:
                number_of_bedrooms_needed = 2
        else: # alternative partnered branch
            number_of_bedrooms_needed = 3

    return number_of_bedrooms_needed

# function trials driving towards different branch alternatives

print(desired_bedrooms(2,True,30000)) # Expected 1
print(desired_bedrooms(2,False,30000)) # Expected 2
print(desired_bedrooms(2,True,50000)) # Expected 3
print(desired_bedrooms(2,False,50000)) # Expected 3
print(desired_bedrooms(3,True,30000)) # Expected 1
print(desired_bedrooms(4,True,30000)) # Expected 3
print(desired_bedrooms(5,False,30000)) # Expected 2
print(desired_bedrooms(5,False,90000)) # Expected 4
print(desired_bedrooms(5,False,80000)) # Expected 4
```
