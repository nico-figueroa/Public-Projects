```python
#This function evaluates if the input number is between 3 and 18 while also 
# being an odd number and returns a boolean value with the result
def is_odd_stat(whole_number):
    odd_result = (3 <= whole_number <= 18) and ((whole_number % 2) != 0)
    return odd_result
```

```python
#This function receives any word as an argument and returns a boolean value
# describing the result of reviewing whether the word starts with the letter i 
# or ends with the letter n.
def has_i_or_n(input_word):
    start_with_i_or_end_with_n = input_word.startswith(("i","I")) or input_word.endswith(("n","N")) # checks for both lower and upper case
    return start_with_i_or_end_with_n
```

```python
has_i_or_n(word)
```

```python
#This function evaluates if both De Morgan laws are true for two boolean values entered as arguments
def de_morgan(P,Q):
    and_de_morgan = (not (P and Q)) == ((not P) or (not Q)) #evaluation of first De Morgan law
    or_de_morgan = (not (P or Q)) == ((not P) and (not Q)) #evaluation of second De Morgan law 
    both_de_morgan_confirmed = and_de_morgan and or_de_morgan # confirmation that both laws hold true
    return both_de_morgan_confirmed #returns the evaluation as a boolean value
```

```python
print(de_morgan(False,False))
print(de_morgan(True,False))
print(de_morgan(False,True))
print(de_morgan(True,True))
```
