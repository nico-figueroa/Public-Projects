```python
import pandas as pd
import numpy as np
```

```python
grants_df = pd.read_csv('data/BMGF-Grants-july2023.csv')
print(len(grants_df))
```

```python
print(list(grants_df.columns))
```

```python
grants_df.head(5)
```

```python
largest_grant = grants_df.iloc[grants_df['AMOUNT COMMITTED'].idxmax()]
print(largest_grant)
```

```python
grants_by_year = grants_df.groupby('YEAR')
```

```python
amount_committed_per_year = grants_by_year['AMOUNT COMMITTED'].sum()
print(amount_committed_per_year)
```

```python
yearly_statistics_df = grants_by_year.agg(
    total=('AMOUNT COMMITTED','sum'),
    median=('AMOUNT COMMITTED','median'),
    maximum=('AMOUNT COMMITTED','max')
)
print(yearly_statistics_df)
```

```python
total_amount_per_region = grants_df.groupby('REGION SERVED').agg({'AMOUNT COMMITTED':'sum'})['AMOUNT COMMITTED']
print(total_amount_per_region)
```

```python
number_grants_per_region_df = grants_df.groupby('REGION SERVED').agg(COUNT=('GRANT ID','count')).reset_index('REGION SERVED')
print(number_grants_per_region_df)
```

```python
print(number_grants_per_region_df['COUNT'].sum())
```

```python
regionless_df = grants_df.loc[(grants_df[['REGION SERVED']].isnull()).any(axis='columns')]
print(regionless_df[['GRANT ID','GRANTEE','YEAR']])
```

```python
top_5_grantees_df = grants_df.groupby('GRANTEE').agg({'AMOUNT COMMITTED':'sum'}).sort_values(by=['AMOUNT COMMITTED'],ascending=False).reset_index('GRANTEE').head(5)
print(top_5_grantees_df)
```

```python
amounts_by_year_and_topic = grants_df.groupby(['YEAR','TOPIC']).agg({'AMOUNT COMMITTED':'sum'}).sort_values(by=['AMOUNT COMMITTED'],ascending=False)
most_funded_topics_df = amounts_by_year_and_topic.loc[(amounts_by_year_and_topic.groupby('YEAR').agg(INDEX=('AMOUNT COMMITTED','idxmax')))['INDEX']]
print(most_funded_topics_df)
```
