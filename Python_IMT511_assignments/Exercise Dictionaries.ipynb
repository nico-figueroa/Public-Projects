```python
# The get area function takes a circle or rectangle shape as an argument
# and returns the area of the shape.

import math

def get_area(shape):
    
    if 'radius' in shape:
    
        shape_area = math.pi*shape['radius']**2 # Area of a circle = π x r ^ 2
    
    else:
    
        shape_area = shape['width']*shape['height'] # Area of a rectangle = width x height

    return shape_area
```

```python
print(get_area(rectangle))
print(get_area(circle))
```
