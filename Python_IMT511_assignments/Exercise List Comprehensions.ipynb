```python
tongue_twister = "how much wood would a woodchuck chuck if a woodchuck could chuck wood"
words = tongue_twister.split(sep = ' ')
print(words)
```

```python
initials = [word[0].upper() for word in words]
print(initials)
```

```python
long_words = [word for word in words if len(word) > 3]
print(long_words)
```

```python
long_initials = [word[0].upper() for word in words if len(word) > 3]
print(long_initials)
```

```python
dice_rolls = [(first_dice,second_dice) for first_dice in range(1,7) for second_dice in range(1,7)]
print(dice_rolls)
```

```python
craps_rolls = [craps for craps in dice_rolls if (craps[0] + craps[1]) == 7 or (craps[0] + craps[1]) == 11]
print(craps_rolls)
```
