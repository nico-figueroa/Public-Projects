```python
deck = []
for suit in ["hearts", "diamonds", "clubs", "spades"]:
    for rank in range(2, 14+1):
        deck.append((rank, suit))
```

```python
# The contains_queen_of_hearts function receives a list of cards as tuples as an argument
# and returns True if the Queen of Hearts is in the list

def contains_queen_of_hearts(list_of_cards):

    contains_queen_of_hearts = False # initialize the variable to return

    for card in list_of_cards: # search through all cards
        if card == (12,'hearts'): # look for queen of hearts
            contains_queen_of_hearts = True # if found, return True

    return contains_queen_of_hearts
```

```python
print(contains_queen_of_hearts(deck[10:15]))
print(contains_queen_of_hearts(deck[20:25]))
```

```python
# The get_high_card function receives a list of cards as tuples as an argument
# and returns the card (tuple) with the highest rank.

def get_high_card(list_of_cards):

    high_card = list_of_cards[0] # start with the first element

    for card in list_of_cards: # check every card in the list
        if card[0] > high_card[0]: # perform king of the hill challenge
            high_card = card # assign the new king to the high rank

    return high_card
```

```python
test_deck = [deck[0],deck[15],deck[25],deck[40],deck[51]]
print(get_high_card(test_deck))
```

```python
# The is_flush function receives a list of cards as tuples as an argument
# and returns True if all cards are of the same suit.

def is_flush(list_of_cards):

    is_flush = True # assume all cards are the same suit

    for card in list_of_cards:
        
        if card[1] != list_of_cards[0][1]: # if the suit of the card is different than the suit of the first card, its not a flush
           
            is_flush = False # change the variable to False
            break # no need to search further

    return is_flush
```

```python
print(is_flush(deck[0:5]))
print(is_flush(deck[12:17]))
```

```python
# The get_face_cards function receives a list of cards as tuples as an argument
# and returns a new list containing only the tuples representing the face cards
# (Jack, Queen or King).

def get_face_cards(list_of_cards):

    # filtering list comprehension that iterates through all cards and return those matching 11, 12 and 13 (Jack, Queen or King)
    face_cards = [card for card in list_of_cards if card[0] == 11 or card[0] == 12 or card[0] == 13] 

    return face_cards
```

```python
print(get_face_cards(deck))
```

```python
# The has_pair function receives a list of cards as tuples as an argument
# and returns a whether or not there is at least one pair in the list and
# the value of the pair that contains the highest rank.

def has_pair(list_of_cards):

    pair_found = False # assumes there is no pair
    pair_value = -1 # when there is no pair, there is no pair value
    
    for i in range(len(list_of_cards)):  # pick the first card in the list      
        for j in range(i+1,len(list_of_cards)): # iterate through all other cards in the list

            if list_of_cards[i][0] == list_of_cards[j][0]: # check for a pair between both cards
                pair_found = True 
                if pair_value < list_of_cards[i][0]: # check if the recent pair found is the highest (king of the hill)
                    pair_value = list_of_cards[i][0]

    return pair_found, pair_value

test_deck = [(2,'hearts'),(14,'hearts'),(11,'spades'),(6,'spades'),(6,'hearts'),(13,'diamonds'),(2,'diamonds'),(12,'spades')]
print(has_pair(test_deck))
```
