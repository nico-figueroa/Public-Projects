```python
with open('data/words_small.txt') as small_words_file:
    words_str = small_words_file.read()
    print(words_str[0:100])
```

```python
count = 0
for string in words_str:
    if string == "\n":
        count = count + 1
print(count)

print(words_str.count("\n"))
```

```python
letter_counts = {}

for letter in words_str:
    if letter != "\n":
        if letter not in letter_counts.keys():
            letter_counts.update({letter:1})
        else:
            letter_counts[letter] = letter_counts.get(letter) + 1

print(letter_counts)
```

```python
for key in sorted(letter_counts.keys()):
    print(key.upper(), "appears",letter_counts[key], "times")
```

```python
for item in letter_counts.items():
    if item[1] == 515:
        print(item[0])
```

```python
count_to_letter_dict ={}
for key, value in letter_counts.items():
    count_to_letter_dict.update({value : key})
print(count_to_letter_dict)
```

```python
for key in (sorted(count_to_letter_dict.keys(),reverse=True))[0:10]:
    print(count_to_letter_dict[key].upper(), "appears", key, "times")
```

```python
with open('data/words_huge.txt') as huge_words_file:
    words_str_huge = huge_words_file.read()

letter_counts_huge = {}

for letter in words_str_huge:
    if letter != "\n":
        if letter not in letter_counts_huge.keys():
            letter_counts_huge.update({letter:1})
        else:
            letter_counts_huge[letter] = letter_counts_huge.get(letter) + 1

count_to_letter_dict_huge ={}
for key, value in letter_counts_huge.items():
    count_to_letter_dict_huge.update({value : key})

for key in (sorted(count_to_letter_dict_huge.keys(),reverse=True))[0:10]:
    print(count_to_letter_dict_huge[key].upper(), "appears", key, "times")
```
