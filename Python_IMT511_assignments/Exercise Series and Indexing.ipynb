```python
words = pd.Series(["the", "quick", "brown", "fox", "jumped", "over", "lazy", "dog"])
print(words)
```

```python
capital_words = words.str.capitalize()
print(capital_words)
```

```python
words_a_f = words[(words.str[0].str.lower() >= 'a') & (words.str[0].str.lower() <= 'f')]
print(words_a_f)
```

```python
# The word_bin function receives a series of strings and two letters as strings, and returns
# a new series containing all the values from the argument series that fall between those two
# letters alphabetically

def word_bin(string_series,first_letter,second_letter):
    
    strings_in_range = string_series[
        (string_series.str[0].str.lower() >= first_letter.lower()) & # after or including the first letter. Lowercase both sides to ensure match.
        (string_series.str[0].str.lower() <= second_letter.lower())] # before or including the second letter. Lowercase both sides to ensure match.

    return strings_in_range
```

```python
print(word_bin(words,'e','q'))
```
